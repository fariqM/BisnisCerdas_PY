```python
df = pd.read_csv('/content/gdrive/MyDrive/Bisnis_Cerdas/Tugas4/Blooddonorprediction/dataset.csv', na_values = {'?'})
df.head(10)
```

```python
X = []
Y = []
list_clf = []
```

```python
df = df.values
```

```python
X = df[:, :(df.shape[1]-1)]
Y = df[:, df.shape[1]-1]
```

```python
kf = KFold(n_splits=5)
kf.get_n_splits(X)
index = 0
AllAccuracy = ["","","", "", ""]
AllBestModel = ["","","", "", ""]
model = HyperoptEstimator(classifier=any_classifier('cla'), preprocessing=any_preprocessing('pre'), algo=tpe.suggest, max_evals=50, trial_timeout=30)

for train_index, test_index in kf.split(X):
  X_train, X_test = X[train_index], X[test_index]
  Y_train, Y_test = Y[train_index], Y[test_index]
  # perform the search
  model.fit(X_test, Y_test)
  # summarize performance
  acc = model.score(X_train, Y_train)
  AllAccuracy[index] = acc
  AllBestModel[index] = model.best_model()
  index += 1



```

```python
for n in range(5):
  print(AllAccuracy[n])
  print(AllBestModel[n],"\n")
```
